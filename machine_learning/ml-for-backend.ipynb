```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Create the DecisionTreeRegressor with enhanced settings
decision_tree = DecisionTreeRegressor(
    random_state=42,
    criterion='squared_error',
    max_depth=None,
    min_samples_split=5,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    max_leaf_nodes=None
)

# Create the Multioutput Regressor
mor = MultiOutputRegressor(decision_tree)

# Train the regressor
dmor = mor.fit(X_train.copy(), y_train.copy())

# Generate predictions for testing data
y_pred = dmor.predict(X_test.copy())

# Initialize lists to store MSE and MAE for each output
mse_scores = []
mae_scores = []

# Calculate MSE and MAE for each output
for i in range(y_test.shape[1]):
    mse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i])
    mae = mean_absolute_error(y_test.iloc[:, i], y_pred[:, i])
    mse_scores.append(mse)
    mae_scores.append(mae)

# Print MSE and MAE for each output
print("MSE FOR EVERY OUTPUT")
for i, mse in enumerate(mse_scores):
    print(f'MSE for regressor {i + 1}: {mse}')

print("MAE FOR EVERY OUTPUT")
for i, mae in enumerate(mae_scores):
    print(f'MAE for regressor {i + 1}: {mae}')
```
